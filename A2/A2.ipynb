## Задача 2

```python
dataset = load_dataset('anscombe')
literals = ["I", "II", "III", "IV"]
data = []
for l in literals:
    data.append((dataset[dataset['dataset'] == l]["x"].to_numpy(), dataset[dataset['dataset'] == l]["y"].to_numpy()))
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 9))
for i in range(4):
    h = int(i/2)
    w = i%2
    x = data[i][0]
    y = data[i][1]
    axs[h][w].scatter(x, y, c="b")
    axs[h][w].set_title("Dx={:.0f}, Dy={:.2f}, r={:.2f}".format(np.var(x), np.var(y), np.corrcoef(x, y)[0,1]))
```

## Задача 4

```python
# Просто сгенерируем прямую с шумом
xs = np.linspace(0, 10, 25)
a, b = (7.77, 3.33)
ys = xs * a + b + np.random.normal(scale=1.77, size=(25,))

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
prep_xs = xs.reshape(-1,1)

lin_reg.fit(xs.reshape(-1,1), ys)
a_exp = lin_reg.coef_[0]
b_exp = lin_reg.intercept_
print("a_exp = ", a_exp, "  delta = {:.2f}%".format(abs(a_exp - a)/a*100))
print("b_exp = ", b_exp, "  delta = {:.2f}%".format(abs(b_exp - b)/b*100))

y_exp = xs * a_exp + b_exp
plt.scatter(xs, ys)
plt.plot(xs, y_exp, c="g")
plt.show()
```
